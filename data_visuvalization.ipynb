```python
import csv
import matplotlib.pyplot as plt

with open("student-por.csv") as f:
    reader = csv.DictReader(f)

    grade={"A":0, "B":0, "C":0, "D":0}

    for row in reader:
        currentGrade= int(row["G3"])

        if(currentGrade >= 18):
            grade["A"] =  grade["A"]+1
        elif(currentGrade >= 14 and currentGrade < 18):
            grade["B"] = grade["B"]+1
        elif(currentGrade >= 10 and currentGrade < 14):
            grade["C"] = grade["C"]+1
        else:
            grade["D"] = grade["D"]+ 1

    #Plot
    plt.bar(grade.keys(), grade.values(), color="skyblue")
    plt.title("Grade Distribution")
    plt.xlabel("Grades")
    plt.ylabel("No.of student")
    plt.show()
        
        
```

```python
import csv
salesByRegion = {"East":0, "West":0, "North": 0, "South": 0}
salesCountByRegion = {"East":0, "West":0, "North": 0, "South": 0}
with open("superstore.csv",encoding="utf-8") as sd:
    reader = csv.DictReader(sd)

   
    for row in reader:
        sale = float(row["Sales"])
        region = row["Region"]
        if(region in salesByRegion):
           salesByRegion[region] = salesByRegion[region]+  sale
           salesCountByRegion[region] = salesCountByRegion[region]+  1        
    
    # Print totals and averages
    for region in ["East", "West", "North", "South"]:
        average = salesByRegion[region] / salesCountByRegion[region] if salesCountByRegion[region] > 0 else 0
        print(f"Total sales for {region}: {salesByRegion[region]:.2f}")
        print(f"Average sale for {region}: {average:.2f}")

# Top performing region
    print("Top", max(salesByRegion, key=salesByRegion.get))
            

# Plot
plt.bar(salesByRegion.keys(), salesByRegion.values(), color="orange")
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales ($)")
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

salesBySegment = {}

with open("superstore.csv", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        segment = row["Segment"]
        sale = float(row["Sales"])
        salesBySegment[segment] = salesBySegment.get(segment, 0) + sale

# Print totals
for seg, total in salesBySegment.items():
    print(f"Total sales for {seg}: {total:.2f}")

# Top segment
top_segment = max(salesBySegment, key=salesBySegment.get)
print(f"Top segment: {top_segment}")

# Plot
plt.bar(salesBySegment.keys(), salesBySegment.values(), color="teal")
plt.title("Total Sales by Segment")
plt.xlabel("Segment")
plt.ylabel("Sales ($)")
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

sales_by_region = {}
sales_by_segment = {}

with open("superstore.csv", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        region = row["Region"]
        segment = row["Segment"]
        sale = float(row["Sales"])

        sales_by_region[region] = sales_by_region.get(region, 0) + sale
        sales_by_segment[segment] = sales_by_segment.get(segment, 0) + sale

# --- Visualization ---
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Chart 1: Sales by Region
axes[0].bar(sales_by_region.keys(), sales_by_region.values(), color="orange")
axes[0].set_title("Sales by Region")
axes[0].set_xlabel("Region")
axes[0].set_ylabel("Total Sales ($)")

# Chart 2: Sales by Segment
axes[1].bar(sales_by_segment.keys(), sales_by_segment.values(), color="teal")
axes[1].set_title("Sales by Segment")
axes[1].set_xlabel("Segment")

plt.tight_layout()
plt.show()
```
