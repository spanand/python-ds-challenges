```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

data = pd.read_csv("superstore.csv")

#Parse and convert Order Date
format_string = '%d/%m/%Y'
data["Order Date"]=[datetime.strptime(d, format_string) for d in data["Order Date"]]

# ---- Extract Year and Month ----
data["Year"] = data["Order Date"].dt.year
data["Month"] = data["Order Date"].dt.to_period("M")

sale_by_year=data.groupby("Year")["Sales"].sum().reset_index()

monthly_sales=data.groupby("Month")["Sales"].sum().reset_index()


# ---- Plot ----
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales["Month"].astype(str), monthly_sales["Sales"], color="teal", marker="o")
plt.title("Monthly Sales Trend (Superstore Data)")
plt.xlabel("Month")
plt.ylabel("Total Sales ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(sale_by_year["Year"].astype(str), sale_by_year["Sales"], color="teal", marker="o")
plt.title("Yearly Sales Trend (Superstore Data)")
plt.xlabel("Year")
plt.ylabel("Total Sales ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
