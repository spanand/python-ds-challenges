```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("superstore.csv", encoding="utf-8")

#clean
data = data.drop_duplicates()
data = data.dropna(subset=["Postal Code"])

#Total and average sales by Category
sales_by_category= data.groupby("Category")["Sales"].agg(["sum","mean"]).reset_index()
sales_by_sub_category= data.groupby("Sub-Category")["Sales"].agg(["sum","mean"]).head().reset_index()
sales_by_segment= data.groupby("Segment")["Sales"].agg(["sum","mean"]).reset_index()

print("Category", sales_by_category)
print("SubCategory", sales_by_sub_category)
print("Segment", sales_by_segment)

# ---- Plot ----
plt.figure(figsize=(6, 6))
plt.bar(sales_by_category["Category"], sales_by_category["sum"], color="teal")
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales ($)")
plt.show()

#Additional challenges
plt.figure(figsize=(6, 6))
plt.bar(sales_by_sub_category["Sub-Category"], sales_by_sub_category["sum"], color="teal")
plt.title("Total Sales by SubCategory")
plt.xlabel("SubCategory")
plt.ylabel("Total Sales ($)")
plt.show()

plt.figure(figsize=(6, 6))
plt.bar(sales_by_segment["Segment"], sales_by_segment["sum"], color="teal")
plt.title("Total Sales by Segment")
plt.xlabel("Segment")
plt.ylabel("Total Sales ($)")
plt.show()

#horizontal bar
plt.figure(figsize=(6, 6))
plt.barh(sales_by_segment["Segment"], sales_by_segment["sum"], color="teal")
plt.title("Total Sales by Segment")
plt.xlabel("Segment")
plt.ylabel("Total Sales ($)")
plt.show()
    
```
