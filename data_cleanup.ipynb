```python
import csv
missing_counts={}
with open("superstore.csv", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    columns = reader.fieldnames
    missing_counts = {col: 0 for col in columns}

    for row in reader:
       for col in columns:
           if(row[col] == "" or "None"):
               missing_counts[col] += 1
        # Print summary
    for col, count in missing_counts.items():
        if count > 0:
            print(f"{col}: {count} missing values")

print("✅ Missing data check complete!")
       
        
            
        
    
```

```python
import pandas as pd

# Load data
df = pd.read_csv("superstore.csv")

# Check for missing values per column
missing_values = df.isnull().sum()

# Display columns with any missing values
print("Missing Values Summary:")
print(missing_values[missing_values > 0])

duplicates = df.duplicated().sum()
print(f"Total duplicate rows: {duplicates}")

df = df.drop_duplicates()
print("✅ Duplicates removed.")

df = df.dropna(subset=["Postal Code"])
print("✅ Rows with missing Postal Code removed.")


print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# ✅ Basic Summary
print("Dataset Summary:")
##print(df.describe())

print("\nTop 5 Categories by Sales:")
print(df.groupby("Category")["Sales"].sum().sort_values(ascending=False).head())

print("\nTop 5 Sub-Categories by Sales:")
print(df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head())

# ✅ Visualization 1: Sales by Category
df.groupby("Category")["Sales"].sum().plot(kind="bar", color="teal")
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales ($)")
plt.show()

# ✅ Visualization 2: Sales by Region
df.groupby("Region")["Sales"].sum().plot(kind="bar", color="orange")
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales ($)")
plt.show()
```
